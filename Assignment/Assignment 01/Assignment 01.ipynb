```python
import csv
```

```python
rows = []
with open('titanic-train.csv', 'r') as file:
    csvReader = csv.reader(file)
    header = next(csvReader)
    for row in csvReader:
        rows.append(row)

print(header)
print(rows)
print(len(rows))
```

```python
# plan to calculate the accuracies
# - for each feature, split them up into separate buckets depending on whether it's a 0 or 1, this can be achieved using a simple for loop
# and if statement
# - in each bin, count the number of people who survived and those who did not survive; this can be achieved using a for loop, a variable
# for survived and not survived, and an if statement
```

```python
# the class feature; is row[0]
bin0_class = []
bin1_class = []
for row in rows:
    if row[0] == '0':
        bin0_class.append(row)
    else:
        bin1_class.append(row)

bin0_class_survive = 0
bin0_class_notsurvive = 0

for bin in bin0_class:
    if bin[-1] == '1':
        bin0_class_survive += 1
    else:
        bin0_class_notsurvive += 1

bin1_class_survive = 0
bin1_class_notsurvive = 0

for bin in bin1_class:
    if bin[-1] == '1':
        bin1_class_survive += 1
    else:
        bin1_class_notsurvive += 1

majority_bin0_class = max(bin0_class_survive, bin0_class_notsurvive)
majority_bin1_class = max(bin1_class_survive, bin1_class_notsurvive)
accuracy = (majority_bin0_class + majority_bin1_class) / (bin0_class_survive + bin0_class_notsurvive + bin1_class_survive + bin1_class_notsurvive)
error = 1 - accuracy
print(error)
```
